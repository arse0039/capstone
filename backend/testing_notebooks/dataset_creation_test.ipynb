Create two data sets from the files. The validation split represents a percentage of which files should be used to train the model and what percentage should be used to validate the model. In this case I used 0.1 so 90% (90 from each directory) will be used to train the model and 10% (10 from each directory) will be used to validate the model.
Output sequence length is set to 480000 to correspond to the clips being 30 seconds in length.
The dataset will comtain the labels paired with the audio files.
It will save two audio sets: train_ds for training and validate_ds for validation
Removed: #output_sequence_length=480000,

```python
def create_data_set(path_name):
    training_ds, validate_ds = tf.keras.utils.audio_dataset_from_directory(
        directory=str(path_name),
        batch_size=64,
        validation_split=0.1,
        seed=0,
        subset='both')
    return training_ds, validate_ds
```

```python
train_data, validate_data = create_data_set('Data')


label_names = np.array(train_data.class_names)
print(label_names)
```

```python
def squeeze_audio(audio, label):
    audio = tf.squeeze(audio, axis=-1)
    return audio, label
```

```python
train_data = train_data.map(squeeze_audio, tf.data.AUTOTUNE)
validate_data = validate_data.map(squeeze_audio, tf.data.AUTOTUNE)
```
